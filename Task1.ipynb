## Functions involved</br>
-- **sigmoid(z)**: A function that takes in a Real Number input and returns an output value between 0 and 1.</br> Meihui

-- **loss(y, y_hat)**: A loss function that allows us to minimize and determine the optimal parameters. The function takes in the actual labels y and the predicted labels yhat, and returns the overall training loss. Note that you should be using the Log Loss function taught in class.</br> Meihui

-- **gradients(X, y, y_hat)**: The Gradient Descent Algorithm to find the optimal values of our parameters. The function takes in the training feature X, actual labels y and the predicted labels yhat, and returns the partial derivative of the Loss function with respect to weights (w) and bias (db). </br> Meihui

-- **train(X, y, bs, epochs, lr)**: The training function for your model.</br>

-- **predict(X)**: The prediction function where you can apply your validation and test sets.




```python
import numpy as np
import pandas as pd
```

```python
train_set=pd.read_csv("train_tfidf_features.csv")
train_set
```

```python
test_set = pd.read_csv("test_tfidf_features.csv")
test_set
```

 $σ(z) =  \frac {1} {1+{e}^{-z}} $

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

$Loss(y,\hat{y}) = -ylog(\hat{y})-(1-y)log(1-\hat{y})$

```python
# TODO: runs out of memory on my laptop, retry ltr
def loss(y, y_hat):
    loss = np.log(1+np.exp((-y)*(y_hat)))
    return loss
```

```python
def loss1(y, y_hat):
    loss = -np.mean(y*(np.log(y_hat)) - (1-y)*np.log(1-y_hat))
    return loss
```

```python
def gradients(X, y, y_hat):
    #w: weight, bias: db
    n = X.shape[0]
    w = (1/n) * np.dot(X.T, (y_hat-y))
    db = (1/n) * np.sum((y_hat-y))
    return w, db
    
```

```python
def train(X, y, bs, epochs, lr):
    theta = np.zeros((X.shape[1],1),dtype=float)
    theta_0 = 0
    
    losses = []
    for _ in range(epochs):
        for i in range(((X.shape[0]-1)//bs) + 1):
            # Batching
            X_batch = X[i*bs : i*bs + bs]
            y_batch = y[i*bs : i*bs + bs].reshape((-1,1))
            
            # Prediction
            y_hat_batch = sigmoid(np.dot(X_batch, theta) + theta_0).reshape((-1,1))
            
            # Gradients
            delta_theta, delta_theta_0 = gradients(X_batch, y_batch, y_hat_batch)
            
            # Update
            theta -= lr * delta_theta
            theta_0 -= lr * delta_theta_0
        
        # Calculate loss for current epoch
        losses.append(loss(y, sigmoid(np.dot(X, theta) + theta_0)))
        
    return theta, theta_0, losses
```

```python
def predict(X, theta, theta_0):
    y_hat = sigmoid(np.dot(X, theta) + theta_0)

    return np.array([1 if h >= 0.5 else 0 for h in y_hat])
```

```python
def accuracy(y, y_hat):
    accuracy = np.sum(y == y_hat) / len(y)
    return accuracy
```

```python
# Main code here
np_train_set = train_set.to_numpy()
np_test_set = test_set.to_numpy()[:,1:]

X = np_train_set[:, 2:]
y = np_train_set[:, 1]

theta, theta_0, losses = train(X[:16000, :], y[:16000], 10, 10, 0.01)

y_hat = predict(np_test_set, theta, theta_0)

pd.DataFrame(y_hat)
```

```python
out = pd.DataFrame(y_hat).set_index(test_set["id"])
pd.DataFrame(np.append(theta, [[theta_0]], axis=0)).to_csv("predictors.csv")
out.to_csv("output.csv")
```
